### Importacion de librerias necesarias

```python
# ================================== PCA COMPLETO (guía en 5 pasos) ==================================
# Usa tu CSV generado: students_synthetic_pca_cluster.csv
# Requisitos: pandas, numpy, matplotlib, seaborn, scikit-learn
# -----------------------------------------------------------------------------------------------
# Si estás en Colab, ajusta CSV_PATH a: "/content/sample_data/students_synthetic_pca_cluster.csv"
# -----------------------------------------------------------------------------------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# -----------------------------------------------------------------------------------------------
# Paso 1. Cargar data
# -----------------------------------------------------------------------------------------------
CSV_PATH = "/content/sample_data/students_synthetic_pca_cluster.csv"  # <- ajusta si es necesario
df_raw = pd.read_csv(CSV_PATH)

# Seleccionar solo numéricas (excluye etiquetas si existieran)
drop_cols = [c for c in df_raw.columns if c.lower() in {"clustertrue", "km_label"} or c.upper().startswith("PCA")]
cols_num = [c for c in df_raw.columns if c not in drop_cols]
df = df_raw[cols_num].select_dtypes(include=["number"]).copy()

print("Dimensiones:", df.shape)
print("Columnas:", list(df.columns))

# -----------------------------------------------------------------------------------------------
# Paso 2. Revisión inicial: nulos, rangos, histogramas, correlación
# -----------------------------------------------------------------------------------------------
print("\n¿Valores nulos por columna?\n", df.isnull().sum())
print("\nResumen estadístico:\n", df.describe().T)

# Histogramas
n = len(df.columns)
ncols = 5
nrows = int(np.ceil(n / ncols))
plt.figure(figsize=(ncols*3.2, nrows*2.6))
for i, c in enumerate(df.columns, 1):
    plt.subplot(nrows, ncols, i)
    plt.hist(df[c].values, bins=20, edgecolor='black')
    plt.title(c, fontsize=8)
    plt.tight_layout()
plt.suptitle("Distribución de variables", y=1.02, fontsize=14)
plt.show()

# Matriz de correlación
plt.figure(figsize=(min(1.0*len(df.columns)+6, 16), min(1.0*len(df.columns)+4, 14)))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm", center=0, square=True)
plt.title("Matriz de correlación")
plt.show()

# -----------------------------------------------------------------------------------------------
# Paso 3. PCA: estandarización, determinante, autovalores/varianza, loadings, n_componentes
# -----------------------------------------------------------------------------------------------
# Estandarizar
scaler = StandardScaler()
Xz = scaler.fit_transform(df.values)

# Determinante de la matriz de correlación (con jitter numérico para evitar singularidades)

R = df.corr().abs()
upper = R.where(np.triu(np.ones(R.shape), k=1).astype(bool))
to_drop = [c for c in upper.columns if any(upper[c] >= 0.95)]
print("Cols a eliminar por alta correlación:", to_drop)

df_red = df.drop(columns=to_drop)
Xz_red = StandardScaler().fit_transform(df_red)
pca_red = PCA().fit(Xz_red)

print("Varianza % (primeras 6):", (pca_red.explained_variance_ratio_[:6]*100).round(2))

# PCA completo
pca = PCA()                 # por defecto n_components = n_features
pca.fit(Xz)
autovalores = pca.explained_variance_
var_ratio = pca.explained_variance_ratio_
acum = np.cumsum(var_ratio)

# Tabla de varianza explicada
var_exp = pd.DataFrame({
    "Componente": [f"C{i+1}" for i in range(len(autovalores))],
    "Autovalor": autovalores,
    "Varianza_%": var_ratio*100,
    "Acumulado_%": acum*100
})
print("\nTabla de varianza explicada (primeras 12 filas):")
print(var_exp.head(12).round(2))

# Scree plot (línea en autovalor = 1)
plt.figure(figsize=(7,4))
plt.plot(range(1, len(autovalores)+1), autovalores, marker='o')
plt.axhline(1.0, color='red', linestyle='--', label='Autovalor = 1 (Kaiser)')
plt.xlabel("Componente"); plt.ylabel("Autovalor"); plt.title("Scree Plot")
plt.legend(); plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

# Número de componentes a retener (regla de Kaiser)
n_retener = max(2, int(np.sum(autovalores > 1.0)))
print(f"\nComponentes retenidos por Kaiser (autovalor>1): {n_retener}")

# Loadings (correlación variable–componente): eigenvectors * sqrt(eigenvalues)
loadings = pca.components_.T * np.sqrt(autovalores)
load_df = pd.DataFrame(loadings[:, :n_retener], index=df.columns, columns=[f"C{i+1}" for i in range(n_retener)])
print("\nMatriz de Componentes (Loadings) – componentes retenidos:")
print(load_df.round(3))

# Heatmap de loadings retenidos
plt.figure(figsize=(min(1.2*n_retener+8, 14), min(0.35*len(df.columns)+4, 18)))
sns.heatmap(load_df, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Loadings (variables vs componentes retenidos)")
plt.tight_layout(); plt.show()

# -----------------------------------------------------------------------------------------------
# Paso 3.1. Asignar nombres significativos a componentes (automático por palabras clave)
# -----------------------------------------------------------------------------------------------
def nombre_componente(vars_top):
    v = " ".join(vars_top).lower()
    score = {
        "Rendimiento/Compromiso Académico": any(k in v for k in [
            "student_avg", "lms", "attendance", "assignments", "project", "completion"
        ]),
        "Bienestar Psicoacadémico": any(k in v for k in [
            "stress", "sleep", "self_efficacy", "social_support", "emotional", "procrastination"
        ]),
        "Contexto y Acceso": any(k in v for k in [
            "socioeconomic", "internet", "part_time", "commute", "tech_integration"
        ]),
        "Entorno Docente/Metodología": any(k in v for k in [
            "teacher_evaluation", "interactive_sessions"
        ]),
        "Habilidades Cognitivas": any(k in v for k in ["critical_thinking"])
    }
    # elegir primera categoría que cumpla, si no, usar variables top
    for k, ok in score.items():
        if ok:
            return k
    return " / ".join(vars_top[:2])

# Top variables por componente y renombrado
top_names = {}
for j in range(n_retener):
    comp = load_df.iloc[:, j].abs().sort_values(ascending=False)
    vars_top = comp.index[:4].tolist()
    top_names[j] = nombre_componente(vars_top)

print("\nNombres sugeridos para los componentes:")
for j, name in top_names.items():
    print(f"C{j+1} → {name}  (top: {', '.join(load_df.iloc[:, j].abs().sort_values(ascending=False).index[:4])})")

# -----------------------------------------------------------------------------------------------
# Paso 4. Visualización de componentes (2D y 3D)
# -----------------------------------------------------------------------------------------------
pca_final = PCA(n_components=n_retener)
Z = pca_final.fit_transform(Xz)  # scores

# Mapear títulos con nombres sugeridos
comp_labels = [top_names[i] if i in top_names else f"C{i+1}" for i in range(n_retener)]
# 2D: C1 vs C2
plt.figure(figsize=(6,5))
plt.scatter(Z[:,0], Z[:,1], alpha=0.6)
plt.xlabel(f"C1: {comp_labels[0]}"); plt.ylabel(f"C2: {comp_labels[1]}")
plt.title("Proyección PCA (C1 vs C2)")
plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

# 3D: C1–C3 (si hay 3+)
if n_retener >= 3:
    from mpl_toolkits.mplot3d import Axes3D  # noqa
    fig = plt.figure(figsize=(7,6))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(Z[:,0], Z[:,1], Z[:,2], alpha=0.7)
    ax.set_xlabel(f"C1: {comp_labels[0]}")
    ax.set_ylabel(f"C2: {comp_labels[1]}")
    ax.set_zlabel(f"C3: {comp_labels[2]}")
    ax.set_title("Proyección PCA 3D")
    plt.tight_layout(); plt.show()

# Gráficos 2D adicionales (si hay 3+)
if n_retener >= 3:
    plt.figure(figsize=(6,5))
    plt.scatter(Z[:,1], Z[:,2], alpha=0.6)
    plt.xlabel(f"C2: {comp_labels[1]}"); plt.ylabel(f"C3: {comp_labels[2]}")
    plt.title("Proyección PCA (C2 vs C3)")
    plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

if n_retener >= 3:
    plt.figure(figsize=(6,5))
    plt.scatter(Z[:,0], Z[:,2], alpha=0.6)
    plt.xlabel(f"C1: {comp_labels[0]}"); plt.ylabel(f"C3: {comp_labels[2]}")
    plt.title("Proyección PCA (C1 vs C3)")
    plt.grid(True, alpha=0.3); plt.tight_layout(); plt.show()

# -----------------------------------------------------------------------------------------------
# Paso 5. Reporte breve para el informe (impresiones clave)
# -----------------------------------------------------------------------------------------------
print("\n--- RESUMEN PARA INFORME ---")
print(f"Determinante(R): {det_R:.6f}")
print("Varianza explicada (%):")
for i in range(n_retener):
    print(f"  C{i+1} ({comp_labels[i]}): {var_ratio[i]*100:.2f}%")
print(f"Acumulado C1..C{n_retener}: {acum[n_retener-1]*100:.2f}%")

print("\nVariables con mayor carga por componente:")
for j in range(n_retener):
    s = load_df.iloc[:, j].abs().sort_values(ascending=False)
    top = s.index[:6].tolist()
    print(f"  C{j+1} ({comp_labels[j]}): {', '.join(top)}")

# -----------------------------------------------------------------------------------------------
# (Opcional) Guardar tablas a CSV
# -----------------------------------------------------------------------------------------------
out_dir = Path("./pca_outputs")
out_dir.mkdir(parents=True, exist_ok=True)
var_exp.to_csv(out_dir / "pca_varianza_explicada.csv", index=False)
load_df.round(4).to_csv(out_dir / "pca_loadings.csv")
scores_df = pd.DataFrame(Z, columns=[f"C{i+1}" for i in range(n_retener)])
scores_df.to_csv(out_dir / "pca_scores.csv", index=False)
print("\nArchivos guardados en:", out_dir.resolve())
```

### Clustering Con K-MEANS

```python
# ================================== CLUSTERING CON K-MEANS ==================================
# Requisitos: pandas, numpy, matplotlib, seaborn, scikit-learn
# =============================================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# ------------------------------ Paso 1: Cargar y preparar data ------------------------------
CSV_PATH = "/content/sample_data/students_synthetic_pca_cluster.csv"
df = pd.read_csv(CSV_PATH)

# Variables numéricas (quitamos etiquetas y columnas auxiliares)
drop_cols = [c for c in df.columns if c.lower() in {"clustertrue", "km_label"} or c.upper().startswith("PCA")]
cols_num = [c for c in df.columns if c not in drop_cols]
X = df[cols_num].values

# Estandarizar
scaler = StandardScaler()
Xz = scaler.fit_transform(X)

# Reducción a PCA (3 componentes para visualización y clustering más estable)
pca = PCA(n_components=3, random_state=42)
Xp = pca.fit_transform(Xz)

print("Varianza explicada por PCA(3):", pca.explained_variance_ratio_.sum()*100, "%")

# ------------------------- Paso 2: Selección de número de clusters --------------------------
inertia = []
silhouette = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, n_init=20, random_state=42)
    labels = km.fit_predict(Xp)
    inertia.append(km.inertia_)
    silhouette.append(silhouette_score(Xp, labels))

# Gráfico del codo
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(k_range, inertia, marker='o')
plt.title("Método del Codo (Inertia)")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Inertia")

# Gráfico de Silhouette
plt.subplot(1,2,2)
plt.plot(k_range, silhouette, marker='o')
plt.title("Índice Silhouette por k")
plt.xlabel("Número de clusters (k)")
plt.ylabel("Silhouette")
plt.tight_layout()
plt.show()

# ---------------------- Paso 3: Ajuste del modelo con k óptimo -------------------------------
k_opt = np.argmax(silhouette) + 2  # +2 porque range empieza en 2
print("k óptimo según Silhouette:", k_opt)

km = KMeans(n_clusters=k_opt, n_init=20, random_state=42)
labels = km.fit_predict(Xp)
df["Cluster"] = labels

print("Métricas finales:")
print("Silhouette:", silhouette_score(Xp, labels))
print("Davies-Bouldin:", davies_bouldin_score(Xp, labels))
print("Calinski-Harabasz:", calinski_harabasz_score(Xp, labels))

# ---------------------------- Paso 4: Visualización -----------------------------------------
# 2D PCA (C1 vs C2)
plt.figure(figsize=(7,6))
sns.scatterplot(x=Xp[:,0], y=Xp[:,1], hue=labels, palette="tab10", alpha=0.7)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],
            c="black", s=200, marker="X", label="Centroides")
plt.title(f"Clusters en PCA (C1 vs C2) | k={k_opt}")
plt.xlabel("C1")
plt.ylabel("C2")
plt.legend()
plt.show()

# 3D PCA
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(Xp[:,0], Xp[:,1], Xp[:,2], c=labels, cmap="tab10", alpha=0.6)
ax.set_title("Clusters en PCA 3D")
ax.set_xlabel("C1")
ax.set_ylabel("C2")
ax.set_zlabel("C3")
plt.show()

# ---------------------------- Paso 5: Interpretación -----------------------------------------
# Medias de variables originales por cluster
means = df.groupby("Cluster")[cols_num].mean().round(2)
print("\nMedias por cluster:\n", means)

# Radar plot para comparar perfiles
from sklearn.preprocessing import MinMaxScaler

scaler_radar = MinMaxScaler()
data_norm = scaler_radar.fit_transform(means.T)
data_norm = pd.DataFrame(data_norm, index=means.columns, columns=[f"Cluster {i}" for i in means.index])

angles = np.linspace(0, 2*np.pi, len(data_norm), endpoint=False).tolist()
angles += angles[:1]

plt.figure(figsize=(8,8))
ax = plt.subplot(111, polar=True)

for col in data_norm.columns:
    values = data_norm[col].tolist()
    values += values[:1]
    ax.plot(angles, values, label=col)
    ax.fill(angles, values, alpha=0.05)

ax.set_thetagrids(np.degrees(angles[:-1]), data_norm.index, fontsize=8)
plt.title("Perfiles normalizados por cluster (Radar)", size=14)
plt.legend(loc="upper right", bbox_to_anchor=(1.2, 1.05))
plt.show()
```
